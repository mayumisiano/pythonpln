#Tuplas
###É voltada para elementos heterogêneos, ou seja, distintos, que podem acoplar tipos de variáveis diferentes (ex: strings e inteiros).

###É imutável

###Criadas com parênteses

```python
t = (7, 'roney', 2, 'rogério',)
print(t)
```

```python
t[0] = 5
#Ocorre um erro pois tuplas são imutáveis
```

##Listas de listas

```python
nota= [['roney',10], ['rogerio', 8], ['paula', 5]]
print(nota)
```

###Dicionários

```python
dic = {}
dic['roney'] = 10
dic['rogerio'] = 8
dic['paula'] = 5
dic['alguem'] = 6.5
print(dic)
```

##Para ver uma nota em particular, basta chamar a chave correspondente que foi criada

```python
dic['roney']
```

```python
print(dic.keys())
```

```python
print(dic.values())
```

```python
type(dic.keys())
```

```python
type(dic.values())
```

##Acessando os elementos da chave do dicionário

```python
list(dic.keys())
```

##O primeiro elemento da lista

```python
list(dic.keys())[0]
```

##Função .items() traz todas as tuplas

```python
list(dic.items())
```

```python
for chave, valor in dic.items():
  print(chave, "teve nota", valor)
```
