##Verificando o tipo das variáveis 

```python
type(nltk.corpus.stopwords.words('portuguese'))
```

```python
from nltk.tokenize import RegexpTokenizer

tokenizer= RegexpTokenizer(r'[a-zA-Z]\w+')
tokens = tokenizer.tokenize(corpus)

nova_lista= []

for token in tokens:
  if token.lower() not in stopwords:
    nova_lista.append(token.lower())

frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common()

#Se o token não estiver na lista de stopwords (definida antes como palavras em)
#português, então ele adiciona a essa lista.
#Depois, a frequência de cada uma dessas palavras é mostrada
```

```python
from nltk.tokenize import RegexpTokenizer

tokenizer= RegexpTokenizer(r'[a-zA-Z]\w+')
tokens = tokenizer.tokenize(corpus)

nova_lista= []

##for token in tokens:
##  if token.lower() not in stopwords:
##    nova_lista.append(token.lower())
#Pode ser escrito como

nova_lista = [token.lower() for token in tokens if token.lower() not in stopwords] 

frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common()
```
